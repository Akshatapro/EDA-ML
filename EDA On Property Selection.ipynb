```python
df_obj_1=df_obj.drop(["Alley", "FireplaceQu","PoolQC", "Fence", "MiscFeature"], axis= 1) 
```

```python
df_obj_1
```

```python
#All vaiables with Missing Values Can be treated with Filling Missing Values using Mode.
```

```python
df_obj_1["MasVnrType"].fillna(df_obj_1["MasVnrType"].mode()[0], inplace= True)
df_obj_1["BsmtQual"].fillna(df_obj_1["BsmtQual"].mode()[0], inplace= True)
df_obj_1["BsmtCond"].fillna(df_obj_1["BsmtCond"].mode()[0], inplace= True)
df_obj_1["BsmtExposure"].fillna(df_obj_1["BsmtExposure"].mode()[0], inplace= True)
df_obj_1["BsmtFinType1"].fillna(df_obj_1["BsmtFinType1"].mode()[0], inplace= True)
df_obj_1["BsmtFinType2"].fillna(df_obj_1["BsmtFinType2"].mode()[0], inplace= True)
df_obj_1["Electrical"].fillna(df_obj_1["Electrical"].mode()[0], inplace= True)
df_obj_1["GarageType"].fillna(df_obj_1["GarageType"].mode()[0], inplace= True)
df_obj_1["GarageFinish"].fillna(df_obj_1["GarageFinish"].mode()[0], inplace= True)
df_obj_1["GarageQual"].fillna(df_obj_1["GarageQual"].mode()[0], inplace= True)
df_obj_1["GarageCond"].fillna(df_obj_1["GarageCond"].mode()[0], inplace= True)
```

```python
# Checking if any null value in data

df_obj_1.isna().sum()
```

```python
# Label Encoding of categorical variables

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df_obj_1_le=df_obj_1.apply(le.fit_transform)
df_obj_1_le
```

```python
# Adding Id Column to categorical variables

df_obj_1_le.insert(0,'Id', data['Id'] )
df_obj_1_le
```
